# Exploratory Data Analysis



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/telco_churn.csv")
df.head()
```

```python
df.info()
df.describe(include="all")
```

The dataset contains a mix of numerical and categorical features describing customer demographics, contract details, and billing behavior.

```python
df["Churn"].value_counts(normalize=True)
```

**Key question: Is churn imbalanced?**

```python
sns.barplot(
    data=df,
    x="Contract",
    y=(df["Churn"] == "Yes").astype(int)
)
plt.title("Churn Rate by Contract Type")
plt.show()
```

```python
sns.boxplot(
    data=df,
    x="Churn",
    y="tenure"
)
plt.title("Tenure Distribution by Churn")
plt.show()
```

```python
sns.histplot(
    data=df,
    x="MonthlyCharges",
    hue="Churn",
    bins=30,
    kde=True
)
plt.show()
```

Early EDA suggests that contract type, tenure, and monthly charges are strong churn indicators.
